### Limpieza de los datos ###

En vista de que los titulos de las columnas son bastante largas, se llevará a cabo el cambio en los campos para mejor compresión con ayuda de una variable temporal denominada **desaparecidos_temporal_renombrados**


```python
desaparecidos_temporal_renombrado={"Estado de la desaparición":"Estado_desaparición",
                                   "Clasificación de la desaparición": "Clasificación",
                                   "Fecha de la desaparición": "Fecha_desaparición",
                                   "Edad al momento de la desaparición": "Edad",
                                   "Sexo del desaparecido": "Sexo",
                                   "País de nacimiento del desaparecido": "País_de_nacimiento",
                                   "Ancestro racial del desaparecido":"Ancestro_racial",
                                   "Estado civil del desaparecido":"Estado_civil",
                                   "Escolaridad del desaparecido":"Escolaridad",
                                   "Pertenencia grupal del desaparecido":"Pertenencia_grupal",
                                   "País donde ocurre la desaparición":"País_desaparicion",
                                   "Municipio donde ocurre la desaparición DANE":"Municipio_desparición",
                                   "Departamento donde ocurre la desaparición DANE":"Departamento_desaparición",
                                   "Contexto":"Contexto"}
```

Un análisis de datos exitosa conlleva a realizar cambios en las variables y campos acorde a las necesidades del cliente, por lo que es necesario poner los nuevos titulos de los campos y los datos para almacenarlos en una nueva variable llamada **desaparecidos_df_copia**

Para confirmar la transacción se presenta el comando **desaparecidos_df_copia.head()**

```python
desaparecidos_df_copia=desaparecidos_df.rename(columns=desaparecidos_temporal_renombrado)
desaparecidos_df_copia.head()
```

El primer acto es verificar si los campos no estan vacios con el comando  **.isnull().sum()** el cual rectifica cada registro y lleva una sumatoria

```python
print(desaparecidos_df_copia.isnull().sum())
```

Se puede observar que ningun dato tiene vacios, esto es porque el dato nulo es considerado un tipo texto **"Sin información"**, por lo tanto a continuación se ejecutará el comando el cual ayudará a reemplazar el término por el valor que más se repite en las columnas **Ancestro_racial, Estado civil, País_de_nacimiento, Escolaridad, Pertenencia, País_desaparicion,  y Departamento_desaparición** ya que se pretende analizar la magnitud de personas desaparecidas con las anterior información.


```python
desaparecidos_df_copia['País_de_nacimiento'] = desaparecidos_df_copia['País_de_nacimiento'].replace("Sin información",desaparecidos_df_copia['País_de_nacimiento'].mode().iloc[0])
desaparecidos_df_copia['Escolaridad'] = desaparecidos_df_copia['Escolaridad'].replace("Sin información",desaparecidos_df_copia['Escolaridad'].mode().iloc[0])
desaparecidos_df_copia['Pertenencia_grupal'] = desaparecidos_df_copia['Pertenencia_grupal'].replace("Sin información",desaparecidos_df_copia['Pertenencia_grupal'].mode().iloc[0])
```

```python
desaparecidos_df_copia['Ancestro_racial'] = desaparecidos_df_copia['Ancestro_racial'].replace('Sin Información',desaparecidos_df_copia['Ancestro_racial'].mode().iloc[0])
desaparecidos_df_copia['Estado_civil'] = desaparecidos_df_copia['Estado_civil'].replace("Sin Información",desaparecidos_df_copia['Estado_civil'].mode().iloc[0])
desaparecidos_df_copia['País_desaparicion'] = desaparecidos_df_copia['País_desaparicion'].replace("Sin Información",desaparecidos_df_copia['País_desaparicion'].mode().iloc[0])
desaparecidos_df_copia['Departamento_desaparición'] = desaparecidos_df_copia['Departamento_desaparición'].replace("Sin Información",desaparecidos_df_copia['Departamento_desaparición'].mode().iloc[0])
```

Acto siguiente es determinar que variables necesitan un cambio de tipo de dato.

Según descripción de los campos de la tabla (Títulos) se observa que el campo **'Fecha_desaparición'** es de tipo texto, posee formato de hora y guión "2020-12-7T00:00:00" los cuales no son necesarios para el análisis.

Con el  comando **.str.split('T').str[0]** se procede a eliminar registros que cuenten con la "T" hasta el final del texto. Por otro lado, con el comando **.str.replace("-","")** busca en el texto el "-" y lo reemplaza con nada, haciendo que el dato quede de la siguiente manera: **2020127**

```python
desaparecidos_df_copia['Fecha_desaparición']=desaparecidos_df_copia['Fecha_desaparición'].str.split('T').str[0]
desaparecidos_df_copia['Fecha_desaparición']=desaparecidos_df_copia['Fecha_desaparición'].str.replace("-","")
```

Adicionalmente, observando los registro, se detalla que algunos tienen **'Fecha_desaparición'** con **'Sin información'**, estos registros no son relevantes en el análisis por lo que se eliminan haciendo que desaparecidos_df_copia solo tenga los registros que cumplan con la condición de no tener
**"Sin información"** en el campo de la Fecha.

```python
desaparecidos_df_copia = desaparecidos_df_copia[desaparecidos_df_copia['Fecha_desaparición'] != "Sin información"]
```

También se resalta que en el campo **'Edad'** es de tipo texto y posee números **(1,15,14)** como texto **(Menor de 1 año y Sin información)**

Por lo tanto se realiza nuevamente el procedimiento anterior, haciendo que **desaparecidos_df_copia** no tenga **Menor de 1 año y Sin información** en su campo **'Edad'.**

```python
desaparecidos_df_copia=desaparecidos_df_copia[desaparecidos_df_copia['Edad']!= "Menor de 1 año" ]
desaparecidos_df_copia=desaparecidos_df_copia[desaparecidos_df_copia['Edad']!= "Sin información" ]
```

Siendo datos que a simple vista parecen números y se registran como texto se debe hacer la conversión de un texto a un número con el comando de Pandas **pd.to_numeric()**

```python
desaparecidos_df_copia['Fecha_desaparición']=pd.to_numeric(desaparecidos_df_copia['Fecha_desaparición'])
desaparecidos_df_copia['Edad']=pd.to_numeric(desaparecidos_df_copia['Edad'])
```

El campo **'Fecha_desaparición'** tiene que ser de tipo entero para que el comando de Pandas **pd.to_datetime() con el formato '%Y%m%d' Lo que equivale a año, mes y día** lo convierta a tipo fecha, es decir, **2020127 = 2020-12-7**

```python
desaparecidos_df_copia['Fecha_desaparición'] = pd.to_datetime(desaparecidos_df_copia['Fecha_desaparición'], format='%Y%m%d')
```

Acto siguiente se llevará a cabo la eliminación de la columna **Contexto** ya que no es necesario para el análisis, por tener un solo valor **Desaparición**

```python
desaparecidos_df_copia = desaparecidos_df_copia.drop(columns=['Contexto'])
```

Por último en la Limpieza de datos se utiliza el comando **dtypes** para rectificar los cambios en los campos incluyendo la desaparición de la columna Contexto, como también que la 'Fecha_desaparición y 'Edad' a tipo fecha **datetime64** y **int64** respectivamente.

## Gráficos y Análisis ##

En la sección  se ejecutará comandos que garantizaran la visualización de gráficos, los cuales se analizaran para mejor comprensión de los interesados

**Cantidad de desapariciones por año**

A continuación se requiere determinar en que años es elevada la magnitud de desapariciones desde el año 1901 al 2024

Por lo que se realizara un diagrama de linea cuyos puntos representaran la cantidad de desapariciones al transcurrir de los años.

- De la variable **desaparecidos_de_copia** en su sección año, se almacena los años de las fechas de las desapariciones.
- Se creara una variable temporal **desapariciones_por_año** y se guardará la cantidad de veces que se repite cada año
-  **plt.figure(figsize=(14, 8))** tamaño de la gráfica
- **plt.plot(desapariciones_por_año.index, desapariciones_por_año.values, marker='o', linestyle='-', color='red', label='Gente desaparecida')** proporciona los datos de la gráfica que son el año y su cantidad, al mismo tiempo se da componentes de diseño como el marker (los puntos), linestyle (linea de estilo) y la leyenda (label), todo ello con un color rojo.
- **plt.fill_between(desapariciones_por_año.index, desapariciones_por_año.values, color='slategrey', alpha=0.3)** garantiza un nitido relleno gris claro en el área debajo de la línea.
- **plt.title('Tendencia de desapariciones através de los años')** Título
- **plt.xlabel('Años')** Leyenda en el eje x
- **plt.ylabel('Cantidad de personas')** Leyenda en el eje y
- **plt.legend()** Leyenda
- **plt.grid(True)** Cuadricula
- **plt.tight_layout()** Ordenar para que nada sobrepase
- **plt.show()** Mostrar gráfica

```python
desaparecidos_df_copia['Año'] = desaparecidos_df_copia['Fecha_desaparición'].dt.year
desapariciones_por_año = desaparecidos_df_copia['Año'].value_counts().sort_index()
plt.figure(figsize=(14, 8)) #Tamaño de gráfica 14 ancho 8 largo
plt.plot(desapariciones_por_año.index, desapariciones_por_año.values, marker='o', linestyle='-', color='red', label='Gente desaparecida')
# marker el punto acorde a la tendencia
plt.fill_between(desapariciones_por_año.index, desapariciones_por_año.values, color='slategrey', alpha=0.3)
#Diseños
plt.title('Tendencia de desapariciones através de los años') #Titulo
plt.xlabel('Años') #Label x
plt.ylabel('Cantidad de personas') #Label Y
plt.legend() #Leyendas
plt.grid(True)
plt.tight_layout() #Orden
plt.show() #Mostrar
```

**Análisis Tendencia de desapariciones por años**

- A lo largo de los años  de 1900 hasta aproximadamente 1970 hay una tendencia general de forma plana en la que se evidencia poco indice de desapariciones.
- A finales de 1970  hasta los años 2005 aproximadamente se presenta un incremento de desapariciones forzosas debido al conflicto armado, narcotrafico y violencia.
- El periodo con más desapariciones registras es desde al año 2005 hasta el 2022 ya que para ese entonces se pronunciaban los grupos paramilitares y su constante violencia en territorios rurales y urbanos.
- Se evidencia una caida pronunciada y significativa en el valor de las desapariciones desde el año del 2020 hasta la actualidad debido al inicio de la pandemia COVID19 como también por las negociaciones de Paz por parte del gobierno y lideres de grupos paramilitares lo cual sugiere la disminución de la actividad criminal.



**Gráfica Paises de nacimiento de los desaparecidos**

Se busca obtener la cantidad de personas que nacieron en Colombia antes de su desaparición, por ello se realiza lo siguiente:

- **plt.figure** Tamaño de gráfico
- **frecuencia_paises** variable que recolectará el nombre de país como clave con sus correspondiente valor de repeticiones en los registros.
- **frecuencia_paises.head(16)** La variable tendra en cuenta solamente los primeros 16 ya que son los que tienen más frencuencia en comparación de todos los paises registrados.
- **ax** variable que contiene las instrucciones de diseño de una gráfica en barra vértical con color de relleno y línea de contorno.
- **Ciclo for** en que se ubicará cada anotación en alto, ancho, posición, rotación, espaciado y tipo de letra.
- **plt.title** Título del gráfico
- **plt.xlabel** Leyenda en eje x
- **plt.ylabel** Leyenda en eje y
- **plt.tight_layout()** Organización sin sobreponerse
- **plt.show()** Mostrar la gráfica



```python
plt.figure(figsize=(10, 15))
frecuencia_paises= desaparecidos_df_copia['País_de_nacimiento'].value_counts()
frecuencia_paises=frecuencia_paises.head(16)
ax = frecuencia_paises.plot(kind='bar', color='cyan', edgecolor='black') #diseño
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 20), textcoords='offset points', rotation=90)

plt.title('Paises de nacimiento de los desaparecidos')
plt.xlabel('Paises')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```

**Análisis de frecuencia de paises de nacimientos de las victimas**

- Se visualiza que los paises con mayor frencuencia en su mayoria pertenecen a continente Américano y pocos son de Europa.
- Acorde a la visualización se declara que la mayoria de las victimas nacieron en Colombia.
- El segundo país con más cantidad es Venezuela, hecho que es por el constante cotacto entre los paises vecinos en cuestión de recursos y habitantes.
- La gráfica presenta una cola larga (long tail) donde muchos paises tienen una frecuencia baja, esto sugiere que fue muy poco las desapariciones al transcurrir de los años.

**Gráfica de Cantidad de desapariciones por Pais**

- **frecuencia_paises.head(12)** La variable tendra en cuenta solamente a los paises de más frecuencia debido a que vienen siendo las que tendrán relevancia a la hora de hacer el análisis.

**Gráfico edad entre hombres y mujeres**

A partir de los datos de edad y sexo, se requiere determinar en que rango de edades se encuentran más vulnerables las victimas Hombres y Mujeres.

**Descripción del código**

-**groups** variable temporal que adquiere los sexos de los desaparecidos y lo agrupa.
- **n_groups** hace el conteo de cuantos grupos de sexo hay (2)
- **fig, axes** las gráficas por numero de grupos 2 filas y 1 columna con sus repectivo tamaño.
- **if n_groups == 1:** se asegura que a pesar de ser un solo grupo se agrega a un array
- El ciclo for garantiza la creación de los gráficos acorde al nombre del grupo y espacios, adicionalmente, proprciona la creación y visualización de los Histograma cuyos parámetros son el campo **'Edad al momento de la desaparición'** agrupado con **groups 'Sexo del desaparecido'**. El **bins** permite la creación de intervalos y se asigna estilo como color y borde.
- También en el ciclo da el titulo de gráfico **ax.set_title**, y de los ejes x y y **ax.set_xlabel - ax.set_ylabel**
- **plt.tight_layout()** hace que los componentes de los gráficos no se sobrepongan.
- **plt.show** muesta la gráfica.


```python
groups = desaparecidos_df_copia.groupby('Sexo')

n_groups = len(groups)

# Crear una figura y ejes
fig, axes = plt.subplots(n_groups, 1, figsize=(8, 6))

# Asegurar que axes sea un array, incluso si hay un solo grupo
if n_groups == 1:
    axes = [axes]

# Dibujar un histograma para cada grupo
for (name, group), ax in zip(groups, axes):
    ax.hist(group['Edad'], bins=15, edgecolor='red', color='yellow')
    ax.set_title(f'Sexo del desaparecido: {name}')
    ax.set_xlabel('Edad al momento de la desaparición')
    ax.set_ylabel('Frecuencia')

# Ajustar el layout
plt.tight_layout()

plt.show()
```

- Se detalla que después de los 25 años a las mujeres se presenta una disminución  de cola larga (long tail) que proclama que las desapariciones en edades avanzadas es poco recurrente.
- Las edades más avanzadas para ambos sexos es cercana o sobrepasada de los 100 años.

```python
frecuencia_escolaridad = desaparecidos_df_copia['Escolaridad'].value_counts()
fig, ax = plt.subplots(figsize=(10, 10))


for i, (label, value) in enumerate(frecuencia_escolaridad.items()):
    ax.scatter([i] * value, range(value), color='black', s=30, edgecolor='skyblue')
    ax.annotate(f'{value}', (i, value - 1), ha='center', va='bottom', xytext=(0, 10), textcoords='offset points')

ax.set_xticks(range(len(frecuencia_escolaridad)))
ax.set_xticklabels(frecuencia_escolaridad.index, rotation=45, ha='right')
plt.title('Escolaridad de los desaparecidos')
plt.xlabel('Escolaridad')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```
